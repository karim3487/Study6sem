Моделирование распределений (ЛР №3)

Задание:
Смоделировать выборку псевдослучайных чисел объёмом n=100 в соответствии c распределением Парето.

```python
alpha = 2
x_m = 1

n = 100
sample = (np.random.pareto(alpha, n) + 1) * x_m

def pareto_pdf(x, alpha, x_m):
    return alpha * (x_m ** alpha) / (x ** (alpha + 1))

def pareto_cdf(x, alpha, x_m):
    return 1 - (x_m / x) ** alpha

x_min = x_m
x_max = np.max(sample) * 1.1
x = np.linspace(x_min, x_max, 1000)


count, bins, _ = plt.hist(sample, 50, density=True, label="Гистрограмма")
# plt.plot(x, pareto_pdf(x, alpha, x_m), label="Плотность распределения")

plt.xlabel('Значение')
plt.ylabel('Вероятность')
plt.title('Распределение Парето')
plt.legend()
plt.show()

plt.plot(x, pareto_cdf(x, alpha, x_m), label="Функция распределения")
plt.xlabel('Значение')
plt.ylabel('Вероятность')
plt.title('Распределение Парето')
plt.legend()
plt.show()

```
